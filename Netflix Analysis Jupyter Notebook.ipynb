# Netflix Movies & TV Shows Data Cleaning and Exploratory Data Analysis

**Tools:** PostgreSQL (SQL), Microsoft Excel

**Dataset:** Netflix Movies & TV Shows (Kaggle)

## 1. Project Objective
The objective of this project is to clean, explore, and analyze the Netflix Movies & TV Shows dataset using SQL. The dataset was cleaned before exploratory data analysis (EDA), and the findings were presented using an interactive Excel dashboard.

## 2. Dataset Understanding
The dataset contains information about Netflix Movies and TV Shows including show_id, type, title, director, cast, country, date_added, release_year, rating, duration, listed_in, and description.

**Primary Key:** show_id

```
-- Display first five records
SELECT * FROM netflix_titles LIMIT 5;

-- Count total records
SELECT COUNT(*) FROM netflix_titles;

-- View table structure
SELECT column_name, data_type
FROM information_schema.columns
WHERE table_name='netflix_titles';
```

## 3. Data Cleaning
Cleaning activities included checking missing values, removing duplicates, standardizing text values, converting date formats, and validating unique show IDs.

```
-- Check missing directors
SELECT * FROM netflix_titles WHERE director IS NULL;

-- Check duplicates
SELECT show_id, COUNT(*)
FROM netflix_titles
GROUP BY show_id
HAVING COUNT(*)>1;
```

## 4. Exploratory Data Analysis
Analyses performed:
- Content Type Distribution
- Titles Added by Year
- Top 5 Genres
- Rating Distribution
- Content Release by Year
- Top 5 Movie Durations

```
-- Content Type Distribution
SELECT type, COUNT(*) AS total
FROM netflix_titles
GROUP BY type;
```

```
-- Titles Added by Year
SELECT EXTRACT(YEAR FROM date_added) AS year, COUNT(*) AS total
FROM netflix_titles
GROUP BY year
ORDER BY year;
```

```
-- Top 5 Genres
SELECT listed_in, COUNT(*) AS total
FROM netflix_titles
GROUP BY listed_in
ORDER BY total DESC
LIMIT 5;
```

```
-- Rating Distribution
SELECT rating, COUNT(*) AS total
FROM netflix_titles
GROUP BY rating
ORDER BY total DESC;
```

```
-- Content Release by Year
SELECT release_year, COUNT(*) AS total
FROM netflix_titles
GROUP BY release_year
ORDER BY release_year;
```

```
-- Top 5 Movie Durations
SELECT duration, COUNT(*) AS total
FROM netflix_titles
WHERE type='Movie'
GROUP BY duration
ORDER BY total DESC
LIMIT 5;
```
